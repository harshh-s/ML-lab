# Heirarchical Clustering

```python
x = fl.loc[:,['Glucose', 'Insulin']]
```

```python
# Visualising the dendrogram
import scipy.cluster.hierarchy as sch

fig = plt.figure(figsize=(25, 10))
dendrogram=sch.dendrogram(sch.linkage(x,method='ward'))
plt.title("Dendrogram")
plt.xlabel("X")
plt.ylabel("Eucledian distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

hiearchical_clustering = AgglomerativeClustering(n_clusters = 2, affinity = 'euclidean', linkage = 'ward')
predicted_hc = hiearchical_clustering.fit_predict(x)
predicted_hc
x['Label'] = predicted_hc
```

```python
plt.scatter(x['Glucose'], x['Insulin'], c = predicted_hc)
plt.xlabel('Glucose')
plt.ylabel('Insulin')
plt.show()
```

```python
plt.scatter(fl['Glucose'], fl['Insulin'], c = fl['Outcome'])
plt.xlabel('Glucose')
plt.ylabel('Insulin')
plt.show()
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN

db_clustering = DBSCAN(eps=3,min_samples=5,metric='euclidean')
predicted_db = db_clustering.fit_predict(x)
predicted_db
```

```python
plt.scatter(x['Glucose'], x['Insulin'], c = predicted_db)
plt.xlabel('Glucose')
plt.ylabel('Insulin')
plt.show()
```

```python
correct = []
for i in range(0,767):
    if x['Label'][i] == fl['Outcome'][i]:
        correct.append(1)
    else:
        correct.append(0)
correct[0:10] # -> if we find correctly 1, if not 0



print("Hierarchical Clustering Accuracy : ", (correct.count(1)/x['Label'].size)*100)
```
